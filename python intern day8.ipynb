```python
# Simple Rule-Based Chatbot

print("Hello! I’m ChatBuddy ")
print("Type 'bye' to end the chat.\n")

while True:
    user_input = input("You: ").lower()

    if "hello" in user_input or "hi" in user_input:
        print("ChatBuddy: Hello there! How can I help you today?")

    elif "how are you" in user_input:
        print("ChatBuddy: I'm just a bot, but I'm doing great! How about you?")

    elif "your name" in user_input:
        print("ChatBuddy: I'm ChatBuddy, your friendly Python chatbot.")

    elif "time" in user_input:
        from datetime import datetime
        print("ChatBuddy:", datetime.now().strftime("The current time is %H:%M:%S."))

    elif "date" in user_input:
        from datetime import datetime
        print("ChatBuddy:", datetime.now().strftime("Today's date is %Y-%m-%d."))

    elif "help" in user_input:
        print("ChatBuddy: I can tell you the time, date, or chat with you casually!")

    elif "bye" in user_input or "exit" in user_input:
        print("ChatBuddy: Goodbye! Have a great day!")
        break

    else:
        print("ChatBuddy: I'm not sure how to respond to that. Try asking something else!")

```
