```python
barkam_portfolio/
├── static/
│   ├── css/
│   │   └── style.css
│   └── images/
│       └── profile.jpg
├── templates/
│   ├── index.html
│   ├── about.html
│   ├── projects.html
│   └── contact.html
├── app.py
├── requirements.txt
└── README.md


from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/about')
def about():
    return render_template('about.html')

@app.route('/projects')
def projects():
    return render_template('projects.html')

@app.route('/contact')
def contact():
    return render_template('contact.html')

if __name__ == '__main__':
    app.run(debug=True)

```
