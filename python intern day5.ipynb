```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# data to be written to the csv file 
data = {
    'date':['2024-01-01','2024-01-01','2024-01-02'] ,
    'Product':['Laptop','Phone','Laptop'],
    'Region':['North','South','East'],
    'Sales':[75000,40000,6000],
    'Quantity':[5,10,4]   }
with open('my_data.csv','w',newline='') as file :
    writer = csv.writer(file)
    writer.writerows(data)
print("CSV file 'my_data.csv' created successfully using csv module.")
csv_file_path='my_data.csv'
print(f"\nData from '{csv_file_path}':")
with open (csv_file_path,mode='r',newline='') as file :
    reader = csv.reader(file)
    for row in reader :
        print(row)
df= pd.read_csv('my_data.csv')
df.head()
df.info
df.describe()
df.isnull().sum()
# total sales
total_sales = df['Sales'].sum()
print("Total Sales:",total_sales)
# sales by product
sales_by_product=df.groupby('Product')['Sales'].sum().sort_values(ascending=False)
print(sales_by_product)
# sales by region
sales_by_region=df.groupby('Region')['Sales'].sum()
print(sales_by_region)
# visualization of data
# Sales by product
sales_by_product.plot(kind='bar',color='skyblue',figsize=(8,5))
plt.title("Total Sales by Product")
plt.ylable('Sales')
plt.show()
# Sales by Region 
sales_by_region.plot(kind='pie',autopct='%1.1f%%',figsize=(6,6))
plt.title("Sales by Region")
plt.ylable("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create the data
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02'],
    'Product': ['Laptop', 'Phone', 'Laptop'],
    'Region': ['North', 'South', 'East'],
    'Sales': [75000, 40000, 60000],
    'Quantity': [5, 10, 4]
}

# ✅ Create a DataFrame
df = pd.DataFrame(data)

# ✅ Save it as a CSV
df.to_csv('my_data.csv', index=False)
print("CSV file 'my_data.csv' created successfully!")

# ✅ Read the CSV file
df = pd.read_csv('my_data.csv')
print("\nPreview of the CSV data:")
print(df.head())

# ✅ Explore data
print("\nInfo:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

# ✅ Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# ✅ Total sales
total_sales = df['Sales'].sum()
print("\nTotal Sales:", total_sales)

# ✅ Group by product
sales_by_product = df.groupby('Product')['Sales'].sum()
print("\nSales by Product:")
print(sales_by_product)

# ✅ Visualization
sales_by_product.plot(kind='bar', color='skyblue', figsize=(6,4))
plt.title('Sales by Product')
plt.ylabel('Total Sales')
plt.show()

sales_by_product.plot(kind='bar', color='skyblue', figsize=(6,4))
plt.title('Sales by Product')
plt.ylabel('Total Sales')
plt.show()
```
