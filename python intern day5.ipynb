```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create the data
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02'],
    'Product': ['Laptop', 'Phone', 'Laptop'],
    'Region': ['North', 'South', 'East'],
    'Sales': [75000, 40000, 60000],
    'Quantity': [5, 10, 4]
}

# ✅ Create a DataFrame
df = pd.DataFrame(data)

# ✅ Save it as a CSV
df.to_csv('my_data.csv', index=False)
print("CSV file 'my_data.csv' created successfully!")

# ✅ Read the CSV file
df = pd.read_csv('my_data.csv')
print("\nPreview of the CSV data:")
print(df.head())

# ✅ Explore data
print("\nInfo:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

# ✅ Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# ✅ Total sales
total_sales = df['Sales'].sum()
print("\nTotal Sales:", total_sales)

# ✅ Group by product
sales_by_product = df.groupby('Product')['Sales'].sum()
print("\nSales by Product:")
print(sales_by_product)

# ✅ Visualization
sales_by_product.plot(kind='bar', color='skyblue', figsize=(6,4))
plt.title('Sales by Product')
plt.ylabel('Total Sales')
plt.show()

sales_by_product.plot(kind='pie',autopct='%1.1f%%',figsize=(6,4))
plt.title('Sales by Product')
plt.ylabel('Total Sales')
plt.show()
```
